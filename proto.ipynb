```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
```

```python
options = webdriver.ChromeOptions()
options.add_argument('headless')
dr = webdriver.Chrome(options=options)
dr.get("https://bandcamp.com/tag/alternative")
WebDriverWait(dr, 0.5)
soup = BeautifulSoup(dr.page_source)
```

```python
r = requests.get("https://bandcamp.com/tag/alternative")
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
x = soup.findAll('a', target='_blank')
[ele.attrs['href'] for ele in x if 'album' in ele.attrs['href'] or 'track' in ele.attrs['href']]
```
